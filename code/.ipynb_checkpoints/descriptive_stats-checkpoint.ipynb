## Are there any patterns in the if we reduce granularity? 
Very often short term data is far too jittery and all that can be observed is white noise. There are several techniques to address the issue and try to understand whether there are underlying patterns in the data.
Aggregation and smoothing are amongst those techniques.

In the next few cells, the data has been aggregated to hourly average per month in order to investigate if there are patterns hidden in the data.

The data in the dataset sums the number of twits every five minutes. It is useful for short term prediction of immediate actions. Such data is similar to that of financial data such as stock prices for short term trade.

Aggregated data will reduce granularity and helps us predict seasonal and long term non-stationary time series. This is more similar to use cases such as forecasting weather patterns.

### Are there any monthly patterns?
We then split the data frame into 12 dataframes, each of which contains data for only one month. Then we aggregate each month's data on hourly basis. Finally we plot the hourly graphs per months for which we do have information.
You might say that we do not need to loop over all months given that we have data only for 3 months and you will be correct. There is really not much extra computational complexity and the code would work if you do have more data.

```python
import calendar
import datetime
hourly_mean = []
hourly_before_standardize = []
df['timestamp'] = pd.to_datetime(df.timestamp)
df.index = pd.to_datetime(df.timestamp)
for i in range(1, 13): #looping over all months
    #setting beginning of month for wach month.
    start_date = '2015-'+str(i)+'-01' 
    
    #setting end of month for each month using calendar library.
    end_date = '2015-'+str(i)+'-'+str(calendar.monthrange(2015,i)[1]) 
    
    #Converting string to datetime, so we can perform comparison
    start_date = datetime.datetime.strptime(start_date, '%Y-%m-%d')
    end_date = datetime.datetime.strptime(end_date, '%Y-%m-%d')
    
    #creating a mask for the dataframe that would filters the range from beginning to end of each month,
    mask = (df['timestamp'] > start_date) & (df['timestamp'] <= end_date)
    mask_df = df.loc[mask]
    
    #Creating an index that we later use for performing aggregation using groupby
    mask_index = mask_df.timestamp

    #Calculating the hourly mean per day of the month and adding the resulting dataframe to a list
    hourly_mean.append(mask_df.groupby(mask_index.index.hour).mean())
    hourly_before_standardize.append(mask_df.groupby(mask_index.index.hour).mean())

```

```python
ncols = 0
for i in range(len(hourly_mean)):
    if len(hourly_mean[i]):
        ncols += 1
fig, axes = plt.subplots(nrows=1, ncols=ncols, figsize=(12,5))


i = 0
for hourly in hourly_mean:
    #Ensuring the dataframe contains data
    if hourly.shape[0]:
        hourly.plot(ax=axes[i])
        i += 1
```

#### Concolution
It seems as there is a month average pattern. February has only few records and perhaps that is why both its range and patters are somehow different.

### Standardizing the Data
- In order to experiment as whether we can reduce jitters in the data we can try to standardize the data.
- Standardization is calculating the standard score of the data using $\Large {z = \frac {(X-\mu)}{\sigma}}$, in which $\mu$ is the mean and $\sigma$ is the standard deviation.
- This results the transforming the data for standard deviation to be 1.

```python
from sklearn.preprocessing import StandardScaler
def standardize(data):
    scaler = StandardScaler() 
    print("BEFORE STANDARDIZING:\nMEAN: {}; STD: {}".format(data.mean(axis=0), 
                                                            data.std(axis=0)))
    data_scaled = scaler.fit_transform(data)
    print("AFTER STANDARDIZING:\nMEAN: {}; STD: {}".format(data_scaled.mean(axis=0), 
                                                           data_scaled.std(axis=0)))
    return data_scaled
```

```python
for i, hourly in enumerate(hourly_mean):
    if len(hourly):
        print("month: {}".format(i+1))
        hourly.value = standardize(hourly.values)
        print()
```

```python
ncols = 0
for i in range(len(hourly_mean)):
    if len(hourly_mean[i]):
        ncols += 1
fig, axes = plt.subplots(nrows=1, ncols=ncols, figsize=(12,5))


i = 0
for hourly in hourly_mean:
    #Ensuring the dataframe contains data
    if hourly.shape[0]:
        hourly.plot(ax=axes[i])
        i += 1
```

```python
print("Comparing February and March:")
print(" - Correlation before standrization: {}".format(pearsonr(hourly_before_standardize[1].value.values, 
                                                             hourly_before_standardize[2].value.values)[0]))
print(" - Correlation after standrization: {}".format(pearsonr(hourly_mean[1].value.values, 
                                                             hourly_mean[2].value.values)[0]))
print()

print(" - RMSE before standardization: {}".format(mean_squared_error(hourly_before_standardize[1].value.values, 
                                                                     hourly_before_standardize[2].value.values)))
print(" - RMSE after standardization: {}".format(mean_squared_error(hourly_mean[1].value.values, 
                                                                     hourly_mean[2].value.values)))
```
